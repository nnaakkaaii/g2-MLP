```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df_raw = pd.read_csv('../results/over_smoothing.csv')
models = ['GAT', 'GCN', 'MLP', 'g2-MLP']
df = df_raw[['num_layers'] + models]
```

```python
df_mean = df.groupby(['num_layers']).agg('mean')
df_mean
```

```python
df_std = df.groupby(['num_layers']).agg('std')
df_std
```

```python
fig = plt.figure(figsize=(15,10))
for k in models:
    plt.errorbar(
        df_mean.index,
        df_mean[k],
        yerr=df_std[k],
        capsize=5,  # エラーバーの横線の長さ
        label=k,
    )
plt.title('Relationship between micro-F1 score on PPI dataset and number of layers.', fontsize=15)
plt.xlabel('number of layers', fontsize=15)
plt.ylabel('micro-f1 score on PPI dataset', fontsize=15)
plt.xticks(df_mean.index, df_mean.index)
plt.ylim(70, 100)
plt.legend()
fig.savefig('../results/over_smoothing.jpg')
plt.close()
```
